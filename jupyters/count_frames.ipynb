```python
import cv2
import csv
import os
import glob
import threading
import mipkit
import matplotlib.pyplot as plt
import numpy as np
```

```python
def count_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    total_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    return int(total_frames)
```

## Training Dataset

```python
load_dir = '/mnt/hdd4T/MERC2020/2020-1//train/*.mp4'
video_file_path = sorted(glob.glob(load_dir))
all_counting_frames = mipkit.multiprocessing.pool_worker(inputs=video_file_path, 
                                                         target=count_frames, 
                                                         num_worker=6)

plt.figure(figsize=(30, 10))
_ = plt.hist(all_counting_frames, bins=np.unique(all_counting_frames))
plt.grid(True)
plt.title('The distribution of numbers of frames in MERC2020-Train')

print('Total videos: ', len(all_counting_frames))
print('The average frames: ', np.average(all_counting_frames))
print('The median frames: ', np.median(all_counting_frames))
```
